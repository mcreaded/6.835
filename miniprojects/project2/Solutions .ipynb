```python
data = []
def mapper(patch_size):
    _map = {"orig_acc":None,"mod_acc":None,"ratio": None}
    _map['ratio']=patch_size
    orig_test=dt.run_tree(ratio=patch_size)
    meth_test = dt.run_tree(ext=1,ratio=patch_size)
    _map['orig_acc']=orig_test
    _map['mod_acc']=meth_test
    return _map
data = [mapper(i) for i in [.1,.2,.3,.4,.5,.6,.7,.8,.9]*12]

d_frm = pd.DataFrame(data)
from pandas.tools.plotting import table
a_x = plt.subplot(111,frame_on=False)
a_x.xaxis.set_visible(False)
a_x.xaxis.set_visible(False)
table(a_x,d_frm)
plt.savefig("dataFrame.png")
```

```python

import six
def render_mpl_table(data, col_width=3.0, row_height=0.625, font_size=14,
                     header_color='#40466e', row_colors=['#f1f1f2', 'w'], edge_color='w',
                     bbox=[0, 0, 1, 1], header_columns=0,
                     ax=None, **kwargs):
    if ax is None:
        size = (np.array(data.shape[::-1]) + np.array([0, 1])) * np.array([col_width, row_height])
        fig, ax = plt.subplots(figsize=size)
        ax.axis('off')

    mpl_table = ax.table(cellText=data.values, bbox=bbox, colLabels=data.columns, **kwargs)

    mpl_table.auto_set_font_size(False)
    mpl_table.set_fontsize(font_size)

    for k, cell in six.iteritems(mpl_table._cells):
        cell.set_edgecolor(edge_color)
        if k[0] == 0 or k[1] < header_columns:
            cell.set_text_props(weight='bold', color='w')
            cell.set_facecolor(header_color)
        else:
            cell.set_facecolor(row_colors[k[0]%len(row_colors) ])
    return ax
```

```python
def plot_saver(batch_size = 6,epochs = 100,latent_dim = 24):
    title = "batch_size="+str(batch_size)+" epochs="+str(epochs)+" latent_dim="+str(latent_dim)
    classic_hist = rnn.run_lstm(batch_size = batch_size,epochs = epochs,latent_dim = latent_dim)
    bid_hist = rnn.run_bid_lstm(batch_size = batch_size,epochs = epochs,latent_dim = latent_dim)
    patches = range(epochs)
    fig, ax = plt.subplots()
    l1,=plt.plot(patches,classic_hist['val_acc'],label="classic")
    l2,=plt.plot(patches,bid_hist['val_acc'],label="bidirec")

    plt.legend(handles=[l1, l2])
    plt.xlabel('epochs')
    plt.ylabel("val_acc")
    plt.title(title)
    plt.savefig("plots/"+title)
```

```python
for i in batch_sizes:
    for d in dims:
        plot_saver(batch_size = i,epochs = epochs,latent_dim = d)
```

```python
def plot_saver(batch_size = 6,epochs = 100,latent_dim = 24):
    title = "batch_size="+str(batch_size)+" latent_dim="+str(latent_dim)
    classic_hist = rnn.run_lstm(batch_size = batch_size,epochs = epochs,latent_dim = latent_dim)
    #bid_hist = rnn.run_bid_lstm(batch_size = batch_size,epochs = epochs,latent_dim = latent_dim)
    patches = range(epochs)
    fig, ax = plt.subplots()
    l1,=plt.plot(patches,classic_hist['acc'],label="train_acc")
    l2,=plt.plot(patches,classic_hist['val_acc'],label="val_acc")

    plt.legend(handles=[l1, l2])
    plt.xlabel('epochs')
    plt.ylabel("accuracy")
    plt.title(title)
    plt.savefig("plots/classic/"+title)
```
